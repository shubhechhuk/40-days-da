```python
import pandas as pd

# Load dataset
df = pd.read_csv("Toyota.csv")

# Drop index column
df.drop(columns=['Unnamed: 0'], inplace=True)

# Convert 'KM' and 'HP' to numeric
df['KM'] = pd.to_numeric(df['KM'].astype(str).str.replace(',', ''), errors='coerce')
df['HP'] = pd.to_numeric(df['HP'], errors='coerce')

# Convert 'Doors' text to numeric
df['Doors'] = df['Doors'].replace({'three': 3, 'four': 4, 'five': 5})
df['Doors'] = pd.to_numeric(df['Doors'], errors='coerce')

# Convert 'FuelType' to categorical
df['FuelType'] = df['FuelType'].astype('category')

# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['KM'].fillna(df['KM'].median(), inplace=True)
df['HP'].fillna(df['HP'].median(), inplace=True)
df['MetColor'].fillna(df['MetColor'].mode()[0], inplace=True)
df['Doors'].fillna(df['Doors'].mode()[0], inplace=True)
df['FuelType'].fillna(df['FuelType'].mode()[0], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(15, 12))

numerical_cols = ['Price', 'Age', 'KM', 'HP', 'CC', 'Weight']

for i, col in enumerate(numerical_cols):
    plt.subplot(3, 2, i+1)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))

for i, col in enumerate(numerical_cols):
    plt.subplot(3, 2, i+1)
    sns.boxplot(x=df[col], color='salmon')
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Price vs FuelType
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='FuelType', y='Price', data=df)
plt.title('Price by FuelType')

# Price vs Automatic
plt.subplot(1, 2, 2)
sns.boxplot(x='Automatic', y='Price', data=df)
plt.title('Price by Transmission (0 = Manual, 1 = Auto)')

plt.tight_layout()
plt.show()
```
